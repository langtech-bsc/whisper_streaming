```python
import os

with open('../export_env_variables_mn.sh') as f:
    for line in f:
        if ('export' in line and "=" in line):
            print(line.strip())
            var, value = line.replace('export ', '', 1).strip().split('=', 1)
            value = value.replace('"', '')
            os.environ[var] = value
```
